```python
Q1 = np.percentile(ch['total day calls'], 25, method='midpoint')
Q2 = np.percentile(ch['total day calls'], 50, method='midpoint')
Q3 = np.percentile(ch['total day calls'], 75, method='midpoint')

print("\nQ1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

print("Upper Limit:", upper_limit)
print("Lower Limit:", lower_limit)

outlier = []

for x in ch['total day calls']:
    if ((x > upper_limit) or (x < lower_limit)):
        outlier.append(x)

print("\nOutliers:")
print(outlier)
```

```python
ch1 = ch['total day calls'] > upper_limit

print("\nOutlier Rows:")
print(ch.loc[ch1])

print("\nOutlier Indexes:")
print(ch.loc[ch1].index)

ch.drop([315, 468, 1057, 1121, 1460, 1869, 2392, 2394], inplace=True)

ch.reset_index(drop=True, inplace=True)

print("\nShape After Removing Outliers:")
print(ch.shape)
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
le = LabelEncoder()

obj = ch.select_dtypes('object')

for i in obj:
    ch[i] = le.fit_transform(ch[i])

print("\nEncoded Data:")
print(ch.head())
```

```python
numeric = ch.drop(
    ['state', 'phone number', 'international plan', 'voice mail plan', 'churn'],
    axis=1
)

apl = ch[['state', 'phone number', 'international plan', 'voice mail plan', 'churn']]

minmax = MinMaxScaler(feature_range=(0,1))

X1 = minmax.fit_transform(numeric)

X1 = pd.DataFrame(X1, columns=numeric.columns)

print("\nScaled Data:")
print(X1.head())
```

```python
X2 = pd.concat([numeric, apl], axis=1)

print("\nCombined Data:")
print(X2.head())
```
